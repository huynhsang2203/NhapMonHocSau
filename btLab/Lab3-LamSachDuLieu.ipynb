```python
column_names=["ID", "Name", "Age", "Weight", "m0006", "m0612","m1218", "f006", "f0612", "f1218"]
df= pd.read_csv('C:\Users\Thai Anh\Downloads\Nhapmonphantichdulieuhocsau\patient_heart_rate.csv', names= column_names)
df
```

```python
df[['Firstname', 'Lastname']]=df['Name'].str.split(expand = True)
df.head(5)
```

```python
df = df.drop('Name',axis=1)
df.head(5)
```

```python
weight=df['Weight']
for i in range(len(weight)):
  x = str(weight[i])
  if "lbs" in x[-3:]:
    #remove the lbs from the value
    x = x[:-3]
    float_x = float(x)
    y = int(float_x/2.205)
    y = str(y)+"kgs"
    weight[i] = y
print(df.head(5))
```

```python
df.dropna(how="all", inplace=True)
df
```

```python
df = df.drop_duplicates(subset=['Firstname','Lastname', 'Age', 'Weight'])
df
```

```python
df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True,inplace=True)
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True,inplace=True)
df
```

```python
index_ageweight = df[df['Age'].isna() & df['Weight'].isna()].index
index_ageweight
```

```python
df.drop(index_ageweight, inplace=True)
df
```

```python
df['Age'].fillna(df['Age'].mean(), inplace =False)
```

```python
df
```

```python
df_ = pd.melt(df, id_vars=['ID', 'Age', 'Weight', 'Firstname', 'Lastname'], value_name="PulseRate",var_name="sex_and_time").sort_values(['ID', 'Age', 'Weight', 'Firstname', 'Lastname'])
```

```python
tmp_df= df_["sex_and_time"].str.extract("(\D)(\d+)(\d{2})", expand=True)
```
